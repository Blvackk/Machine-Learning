```python
def add_sub(x,y): # what if i want to return 2 values add_sub & i want to return 2 values & function can accept multiple value 
    c= x+y
    d= x-y
    
    return c, d 
    
r, r1 = add_sub(4,5)
print(r)
print(r1)

print(type(r))
print(type(r1))
```

```python
def add_sub_mul(x,y): # what if i want to return 2 values add_sub & i want to return 2 values & function can accept multiple value 
    c= x+y
    d= x-y
    e= x*y
    
    return c, d,e
    
r, r1,r2 = add_sub_mul(4,5)
print(r)
print(r1)
print(r2)

print(type(r))
print(type(r1))
```

# kwargs ( kw + args) 

```python
def person():
    person('ALEX', 36, 'JOHN', 987767)
person()
```

```python
def person(name, *data):
    print(name)
    print(data)

person('ALEX', 36, 'JOHN', 987767)
```

```python
def person(name,*data):
    print('name')
    print(data)

person('ALEX', age = 36, home_place ='southcity', mob =987767)
```

```python
def person(name,**data):
    print('name')
    print(data)

person('ALEX', age = 36, home_place ='southcity', mob =987767)
```
